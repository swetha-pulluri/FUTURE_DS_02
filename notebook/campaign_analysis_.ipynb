```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('social_media_campaign.csv')

# Show data
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Engagement_Rate'] = df['Clicks'] / df['Impressions']
df[['Campaign_ID', 'Platform', 'Engagement_Rate']]
```

```python
df['Conversion_Rate'] = df['Conversions'] / df['Clicks']
df[['Campaign_ID', 'Platform', 'Conversion_Rate']]
```

```python
platform_spend = df.groupby('Platform')['Spend'].sum()

platform_spend.plot(kind='bar', title='Spend by Platform')
plt.xlabel('Platform')
plt.ylabel('Total Spend')
plt.tight_layout()
plt.show()
```

```python
plt.bar(df['Campaign_ID'], df['Conversions'])
plt.title('Conversions by Campaign')
plt.xlabel('Campaign ID')
plt.ylabel('Conversions')
plt.tight_layout()
plt.show()
```
